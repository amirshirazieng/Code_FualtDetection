```python
import shutil, os
import pandas as pd
```

```python
train_path = 'Dataset/Training_Set/'
train_csvfilename = 'RFMiD_Training_Labels.csv'


train_df = pd.read_csv(train_path+train_csvfilename)
train_df.head()
```

```python
labels = list(train_df.columns)
for i in labels:
    os.makedirs(os.path.join(train_path, 'Training_Folders/', i))
    
for cd in range(len(labels)):
    counter=0
    for im in range(len(train_df)):
#         print(im+1, cd+1)
        if train_df.iloc[im,cd]==1:
            counter+=1
            im_train_path = os.path.join(train_path,'Training/', str(im+1)+'.png')
#             print(im_train_path)
#             print(os.train_path.join(train_path,'Training_Folders/', labels[cd+1], str(im+1)+'.png'))
            shutil.copy(im_train_path,os.path.join(train_path, 'Training_Folders/', labels[cd], labels[cd]+ '_' +str(counter)+'.png'))
```

```python
test_path = 'Dataset/Test_Set/'
test_csvfilename = 'RFMiD_Testing_Labels.csv'


test_df = pd.read_csv(test_path+test_csvfilename)
test_df.head()
```

```python
labels = list(test_df.columns)
for i in labels:
    os.makedirs(os.path.join(test_path, 'Testing_Folders/', i))
    
for cd in range(len(labels)):
    counter=0
    for im in range(len(test_df)):
#         print(im+1, cd+1)
        if test_df.iloc[im,cd]==1:
            counter+=1
            im_test_path = os.path.join(test_path,'Test/', str(im+1)+'.png')
#             print(im_train_path)
#             print(os.train_path.join(train_path,'Training_Folders/', labels[cd+1], str(im+1)+'.png'))
            shutil.copy(im_test_path,os.path.join(test_path, 'Testing_Folders/', labels[cd], labels[cd]+ '_' +str(counter)+'.png'))
```

```python
eval_path = 'Dataset/Evaluation_Set/'
eval_csvfilename = 'RFMiD_Validation_Labels.csv'


eval_df = pd.read_csv(eval_path+eval_csvfilename)
eval_df.head()
```

```python
labels = list(eval_df.columns)
for i in labels:
    os.makedirs(os.path.join(eval_path, 'Evaluation_Folders/', i))
    
for cd in range(len(labels)):
    counter=0
    for im in range(len(test_df)):
#         print(im+1, cd+1)
        if eval_df.iloc[im,cd]==1:
            counter+=1
            im_eval_path = os.path.join(eval_path,'Validation/', str(im+1)+'.png')
#             print(im_train_path)
#             print(os.train_path.join(train_path,'Training_Folders/', labels[cd+1], str(im+1)+'.png'))
            shutil.copy(im_eval_path,os.path.join(eval_path, 'Evaluation_Folders/', labels[cd], labels[cd]+ '_' +str(counter)+'.png'))
```
